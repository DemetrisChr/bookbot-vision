Image processing alogirthm imported from https://github.com/tphinkle/shelfie

```python
import sys

import numpy as np
import cv2
import matplotlib.pyplot as plt

# Shelfie
sys.path.append('..')
sys.path.append('../models')
import models.image_processing as ip
```

```python
img = cv2.imread('pictures/books1.jpg')

fig = plt.figure(figsize=(16,12))
plt.imshow(img)
plt.show()
```

```python
lines = ip.get_book_lines(img, angles= [-4, 0, 4], debug=False)
```

```python
fig = plt.figure(figsize=(16,12))
plt.imshow(img)

for line in lines:
    centre = line.center
    ys = np.linspace(centre[1]-1000, centre[1]+500, 50)
    xs = list(map(lambda y: line.x(y), ys))
    plt.plot(xs, ys, color='red')
    
centres = list(map(lambda line: line.center, lines))
cen_x = [ c[0] for c in centres ]
cen_y = [ c[1] for c in centres ]
plt.scatter(cen_x, cen_y, marker='x', color='yellow')

plt.show()
```
